```python
living = pd.read_excel('Living Costs.xlsx')
print(living)
living.info()
```

```python
living['INITIAL COST'] = living['INITIAL COST'].astype('str')
living['TOTAL EST. FEE'] = living['TOTAL EST. FEE'].astype('str')
living.info()

    
```

```python
pattern = re.compile('\$\d*')
def diff_money(row, pattern):
    icost = row['INITIAL COST']
    tef = row['TOTAL EST. FEE']
    
    if bool(pattern.match(icost)) and bool(pattern.match(tef)):
        
        icost = icost.replace('$','')
        tef = tef.replace('$','')
        
        icost = int(icost)
        tef = int(tef)
        
        return icost - tef 
    else:
        return(NaN)
    
living['Difference'] = living.apply(diff_money,axis=1,pattern=pattern)

print(living)
      
```
